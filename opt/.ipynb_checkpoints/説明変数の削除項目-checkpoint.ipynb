```python
#三角形の面積を計算する

def calc_tri(base, height):
    return base*height*0.5
```

```python
base = 10
height = "5"

try:
    result = calc_tri(base,height)
    print("計算結果.{}".format(result))
#どんなエラーが起きても引き受けてくれる。
except Exception as e:
    print('計算できません')
    #変数を入れることでエラー文を出力することができる。
    print(e)
```
